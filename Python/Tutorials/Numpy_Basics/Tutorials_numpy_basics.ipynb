```python
def tolowercase(s): # our function name
    # s - input string as parameter
    alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwwxyz" # hold all the alphabets( both caps/small) in a string
             01234567890123456789012345678901234567890123456789012
                       10        20        30        40        50 
    res = ""  # to store the result
    # len of the alphabet string is 52
    
    for ch in s:  # iterate thru all letters of the i/p 
        pos = 0 # postion of string to compare
        
        print(ch)
        while pos < 52:  # to scan thru all the letters of the alpha
            print(pos, alpha[pos], ch)
            if alpha[pos] == ch:  # to compare the existence of a match
                i = pos # if match is found then note that position
            pos = pos + 1
        if ch not in alpha or i>=26:   # if the i/p letter is not in alpha or if the character is already in small letters
            res = res + ch    # keep the i/p character as it is 
        else: # if a match is found then it move 26 index/position to pick the small letter of the same character
            res = res + alpha[i+26]  
    return res
```

```python
ip = " [JAN, FEB, MAR] are the 3-months of the FIRST Quarter !!!"
op = tolowercase(ip)
print('input  string:', ip)
print('output string:', op)
```
